```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/raw/train.csv')
df.sample(3)
```

```python
df.to_csv('train-l.csv')
```

```python
import copy
new_df = copy.deepcopy(df.iloc[:, 1:])
```

```python
new_df = new_df.rename(columns = str.lower)
```

```python
new_df.columns
```

```python
def bivariate_analysis(df, col, x = None):

    cnt =  df[col].value_counts()

    if x is None:
        ax = sns.countplot(data = df, x = col, color = 'purple', hue = 'churn')
    else:
        ax = sns.countplot(data = df, y = col, color = 'purple', hue = 'churn')
        
    plt.xticks(rotation=45) 
    return plt.show()
```

#### Gender vs Churn

```python
bivariate_analysis(new_df, 'gender')
```

#### SeniorCitizen vs Churn

```python
bivariate_analysis(new_df, 'seniorcitizen')
```

```python
bivariate_analysis(new_df, 'partner')
```

```python
bivariate_analysis(new_df, 'dependents')
```
